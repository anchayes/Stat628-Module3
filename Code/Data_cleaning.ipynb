```python
import pandas as pd
import json
json_data = []
json_file = "business.json"
file = open(json_file, encoding='utf-8')
for line in file:
    json_line = json.loads(line)
    json_data.append(json_line)
business = pd.DataFrame(json_data)
business.head()
```

```python
json_data = []
json_file = "review.json"
file = open(json_file, encoding='utf-8')
for line in file:
    json_line = json.loads(line)
    json_data.append(json_line)
review = pd.DataFrame(json_data)
review.head()
```

```python
# merge data
df_raw = pd.merge(review, business, how='left', on='business_id')
df_raw.head()
```

```python
df = df_raw.drop(columns=['useful','funny','cool'])
df.head()
```

```python
# select one supecific restaurants "Breakfast" and remove duplicates lines
Breakfast_index = [i.find("Breakfast")!=-1 for i in map(str,df["categories"])]
Breakfast = df[Breakfast_index]
Breakfast.drop_duplicates(subset='text', keep='first', inplace=True)
Breakfast.shape
```

```python
import string
#remove punctutaion
text = list(Breakfast.text)
text = ["".join([char for char in a if char not in string.punctuation]) for a in text]
text = [i.lower() for i in text]
text[1]
```

```python
import nltk 
nltk.download('words')
nltk.download('punkt')
```

```python
#remove non-English words
Eng_dir = set(nltk.corpus.words.words())
text_Eng = [" ".join([w for w in nltk.word_tokenize(j) if w in Eng_dir or not w.isalpha()]) for j in text]
```

```python
Breakfast['text'] = text_Eng
```

```python
import csv
Breakfast_review = Breakfast.iloc[:,0:10]
Breakfast_review.to_csv("Breakfast_review.csv", index=False)
```
