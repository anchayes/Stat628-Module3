```python
#Import libraries
import json
import numpy as np
import pandas as pd
```

```python
#Open the Yelp Business Data provided, rename it for use, display the top 5 rows
with open("/Users/annahayes/R_stuff/yelp_dataset_2022/business.json", "r") as f:
    df_b = pd.read_json(f, orient="records", lines=True)
df_b.head()
```

```python
#show how many rows the initial business data set contains
len(df_b)
```

```python
#filter the data set to only contain rows that have the words "Breakfast", "Brunch" and combinations in the category
#column, not being case sensative and not including NAS
break_df = df_b[df_b['categories'].str.contains(
              'Breakfast|Brunch|Breakfast&Brunch|Breakfast & Brunch',
              case=False, na=False)]
#display the length of filtered data set
break_df.head()
```

```python
#write the filtered business data set to a csv for user use
break_df.to_csv("yelp_business_breakfast.csv", index=False)
```

```python
#open the review data set for cleaning
with open("/Users/annahayes/R_stuff/yelp_dataset_2022/review.json", "r") as f:
    dfr = pd.read_json(f, orient="records", lines=True)
len(dfr)
```

```python
#Display top 5 rows of data set 
dfr.head()
```

```python
#filter the review data set to only business ids contained in the filtered business data set 
#make data set of just the business ids 
df_id= break_df["business_id"]
len(df_id)
```

```python
#filter to only using reviews that have breakfast business ids 
df_new = dfr.loc[dfr['business_id'].isin(df_id)]
len(df_new)
```

```python
# make a sum column to get reviews that have votes
col_list= list(df_new)
col_list.remove('review_id'); col_list.remove('user_id'); col_list.remove('business_id')
col_list.remove('stars'); col_list.remove('text'); col_list.remove('date')
df2 = df_new.copy()
```

```python
# adding the sum column to the data set
df2.loc[:,'Sum'] = df_new[col_list].sum(axis=1)
```

```python
# filtering the sum column to rows that have 2 or more votes 
df3 = df2[df2['Sum'] >= 6]  
len(df3)
```

```python
#display head of data frame
df3.head()
```

```python
# remove columns we don't need and display head of data frame
df4 = df3.drop(columns=['useful','funny','cool','Sum'])
df4.head()
```

```python
#write the filtered review data set to a csv for user use
df4.to_csv("yelp_review_filtered_breakfast.csv", index=False)
```
